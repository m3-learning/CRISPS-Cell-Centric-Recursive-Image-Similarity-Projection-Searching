```python
# Python code to merge dict using update() method
def Merge(dict_list):
    dict_ = {}
    for i in range(len(dict_list)):
        dict_.update(dict_list[i])
    return dict_
```

```python
from functools import singledispatch


@singledispatch
def keys_to_strings(ob):
    return ob

@keys_to_strings.register
def _handle_dict(ob: dict):
    return {str(k): keys_to_strings(v) for k, v in ob.items()}

@keys_to_strings.register
def _handle_list(ob: list):
    return [keys_to_strings(v) for v in ob]
```

```python
with MPRester('e490iJw7D4f7F0Z0kfNVYKtyWeYGmMM7') as mpr:
    docs = mpr.summary.search(has_props = [HasProps.dielectric, HasProps.piezoelectric])
```

```python
parent_collection = 'c/p_2022_mp_piezoelectric_diffraction_root'

tasklist = []

for doc in docs: 
    try:
        # Sets the original data as a dictionary
        origin = doc.dict()

        # Grabs the materials id
        id = doc.material_id

        props = [origin]


        # extracts the properties
        try:
            elastic = {'elasticitiy': mpr.elasticitiy.get_data_by_id(id).dict()}
            props.append(elastic)
        except: 
            pass
        try:
            piezoelectric = {'piezoelectric': mpr.piezoelectric.get_data_by_id(id).dict()}
            props.append(piezoelectric)
        except: 
            pass
        try:
            dielectric = {'dielectric': mpr.dielectric.get_data_by_id(id).dict()}
            props.append(dielectric)
        except: 
            pass
        try:
            thermo = {'Themo': mpr.thermo.get_data_by_id(id).dict()}
            props.append(thermo)
        except: 
            pass
        try:
            magnetism = {'magnetism': mpr.magnetism.get_data_by_id(id).dict()}
            props.append(magnetism)
        except: 
            pass
        try:
            materials = {'materials': mpr.materials.get_data_by_id(id).dict()}
            props.append(materials)
        except: 
            pass
        try:
            oxi_states = {'oxi_states': mpr.materials.get_data_by_id(id).dict()}
            props.append(oxi_states)
        except: 
            pass
        try:
            bandstructure = {'bandstructure': mpr.materials.get_data_by_id(id).dict()}
            props.append(bandstructure)
        except: 
            pass
        try:
            dos = {'dos': mpr.materials.get_data_by_id(id).dict()}
            props.append(dos)
        except: 
            pass
        try:
            xas = {'xas': mpr.materials.get_data_by_id(id).dict()}
            props.append(xas)
        except: 
            pass
        try:
            electronic_structure = {'electronic_structure': mpr.materials.get_data_by_id(id).dict()}
            props.append(electronic_structure)
        except: 
            pass
        try:
            provenance = {'provenance': mpr.materials.get_data_by_id(id).dict()}
            props.append(provenance)
        except: 
            pass
        try:
            charge_density = {'charge_density': mpr.materials.get_data_by_id(id).dict()}
            props.append(charge_density)
        except: 
            pass
        try:
            substrates = {'substrates': mpr.materials.get_data_by_id(id).dict()}
            props.append(substrates)
        except: 
            pass

        out = Merge(props)

        beam_direction = [[0,0,1],[1,1,1],[0,0,2],[1,0,3],[3,0,1], [6,0,1]]

        for beam_direction_ in beam_direction:
            out['beam_direction'] = beam_direction_
            with MPRester(api_key='e490iJw7D4f7F0Z0kfNVYKtyWeYGmMM7') as mpr:
                # first retrieve the relevant structure
                structure = mpr.get_structure_by_material_id(id)

                # important to use the conventional structure to ensure
                # that peaks are labelled with the conventional Miller indices
                sga = SpacegroupAnalyzer(structure)
                conventional_structure = sga.get_conventional_standard_structure()

                # this example shows how to obtain an XRD diffraction pattern
                # these patterns are calculated on-the-fly from the structure
                calculator = tem.TEMCalculator(beam_direction = beam_direction_)
                pattern = calculator.get_pattern(structure)

                fig = calculator.get_plot_2d(structure)
                fig.write_image(f"figs/{id.string}_{beam_direction_}.png")
                out['ID_name']=f'{id.string}_{beam_direction_}'

                # Makes the data record
                rec_resp = df_api.dataCreate(
                            f'{id.string}_{beam_direction_}',
                            metadata=json.dumps(keys_to_strings(out), default=str),
                            parent_id=parent_collection,
                            )

                 # 2. Extract the record ID from the response
                this_rec_id = rec_resp[0].data[0].id

                # 3. Put the raw data into this record:
                put_resp = df_api.dataPut(this_rec_id, 
                                          f"./figs/{id.string}_{beam_direction_}.png",
                                          wait=False)

                # 4. Extract the ID for the data transfer task
                task_id = put_resp[0].task.id

                tasklist.append(task_id)
                
    except: 
        pass



```

```python
def get_recursively(search_dict, field):
    """
    Takes a dict with nested lists and dicts,
    and searches all dicts for a key of the field
    provided.
    """
    fields_found = []

    for key, value in search_dict.items():

        if key == field:
            fields_found.append(value)

        elif isinstance(value, dict):
            results = get_recursively(value, field)
            for result in results:
                fields_found.append(result)

        elif isinstance(value, list):
            for item in value:
                if isinstance(item, dict):
                    more_results = get_recursively(item, field)
                    for another_result in more_results:
                        fields_found.append(another_result)

    return fields_found
```

```python
def dictionary_check(input):
    """
    First prints the final entry in the dictionary (most nested) and its key
    Then prints the keys leading into this
    * could be reversed to be more useful, I guess
    """
    for key,value in input.items():
        lister = [key]
        if isinstance(value, dict):
            print(key)
            lister.append(key)
            dictionary_check(value)
        else:
            if isinstance(key, str):
                pass
            else:
                print(key)
```
